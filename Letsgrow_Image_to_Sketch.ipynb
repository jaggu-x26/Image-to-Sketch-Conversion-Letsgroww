```python
import cv2
```

```python
# Step 1: Read the image in RGB format
image = cv2.imread('/content/1519826508658.jpg')
```

```python
# Step 2: Convert it to a grayscale image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Step 3: Invert the grayscale image
inverted_gray_image = 255 - gray_image
```

```python
# Step 4: Mix the grayscale image with the inverted blurry image
blurred_image = cv2.GaussianBlur(inverted_gray_image, (21, 21), 0)
pencil_sketch = cv2.divide(gray_image, 255 - blurred_image, scale=256)
```

```python
# Display the original image and the pencil sketch side by side
from google.colab.patches import cv2_imshow
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
